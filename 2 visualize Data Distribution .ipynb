# Box plot

```python
# ดูการคลอส การกระจายของdata
```

```python
import pandas as pd
```

```python
irisdata = pd.read_csv('C:\\Python\\python_dataviz\\iris.data' ,header = None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginia_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [12,12] #ดูการกระจาย
plt.subplot(1,4,1) # มี1แถว 4หลัก ตำแหน่งที่1
plt.boxplot(setosa_np[:,0])
plt.subplot(1,4,2)
plt.boxplot(setosa_np[:,1])
plt.subplot(1,4,3)
plt.boxplot(setosa_np[:,2])
plt.subplot(1,4,4)
plt.boxplot(setosa_np[:,3])
plt.show()
```

```python
plt.boxplot(setosa_np)
```
